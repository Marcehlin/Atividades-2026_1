# Q5

A base de dados decathlon, disponível no pacote FactoMineR do R, reúne informações
de desempenho de atletas de elite que competiram em eventos de decatlo. Cada linha
representa um atleta, e as colunas contêm pontuações ou tempos obtidos nas diferentes
modalidades do decatlo. Para esta análise, foram selecionadas apenas três provas
de corrida: 400 metros rasos (400m), 110 metros com barreiras (110m.hrdle) e 1500
metros rasos (1500m) – todas expressas em tempo (segundos). Essas variáveis permitem
estudar o perfil de velocidade e resistência dos atletas a partir de medidas quantitativas
diretamente ligadas ao desempenho físico. Use o seguinte código no R para carregar os
dados:

```r
library(FactoMineR)
data(decathlon)
```

```r
dim(decathlon)
```

```r
# Selecionar os 10 primeiros atletas e apenas as provas (colunas 1 a 10)
dados_provas <- decathlon[, 1:10]
#stars(dados_sub)
# Gráfico de estrelas

stars(dados_provas,
      draw.segments = TRUE,        # segmentos em vez de linhas poligonais
      key.loc = c(15, 4),          # posição da legenda
      main = "Gráfico de Estrelas - Decathlon (10 primeiros atletas)",
      col.segments = rainbow(10),  # cores para cada variável (opcional)
      nrow = 7, ncol = 6)          # layout da matriz de estrelas
```

```r
# Padronizar (média 0, desvio 1) – mas para estrelas, o melhor é normalizar para [0,1]
normalizar <- function(x) (x - min(x)) / (max(x) - min(x))
dados_norm <- as.data.frame(lapply(dados_provas, normalizar))

# Agora sim, todos os raios variam de 0 a 1 (quanto maior, melhor)
# Atenção: provas de tempo (100m, 400m, 110m com barreiras) precisam ser invertidas:
# pois valores menores são melhores. Vamos inverter:
tempos <- c("X100m", "X400m", "X110m.hurdle")
for (col in tempos) {
  dados_norm[[col]] <- 1 - dados_norm[[col]]
}

# Gráfico com dados normalizados
stars(dados_norm,
      draw.segments = TRUE,
      key.loc = c(15, 5),
      main = "Estrelas - Desempenho normalizado (quanto maior o raio, melhor)",
      col.segments = rainbow(10),
      nrow = 7, ncol = 6)
```

```r
# Ver os rankings
decathlon
# Identificar os três piores (maior número no rank)
piores <- rownames(decathlon[order(decathlon$Rank, decreasing = TRUE), ])[1:3]
print(piores)
```

# Q6
 A base de dados BodyWeight, disponível no pacote nlme do R, contém informações
sobre o peso de ratos medido ao longo do tempo em um experimento com três grupos:

• chow: dieta padrão (controle),

• reduced: dieta com restrição alimentar,

• other: outra dieta alternativa.

Foram observados 16 ratos ao longo de 64 dias, totalizando 176 observações. Cada rato
foi identificado por um código e teve seu peso (em gramas) registrado em diferentes
tempos (em dias). O objetivo do estudo é analisar o efeito da dieta sobre o peso dos
ratos ao longo do tempo. Construa um gráfico de curvas de crescimento e interprete o
efeito das dietas sobre os pesos dos ratos. Use o seguinte código no R para carregar os
dados:


```r
library(nlme)
data(BodyWeight)
print(BodyWeight)
```

```r
cores <- c("red", "blue", "purple")
names(cores) <- levels(BodyWeight$Diet)

plot(range(BodyWeight$Time), range(BodyWeight$weight),
     type = "n",
     xlab = "Tempo",
     ylab = "Peso",
     main = "")

pintinhos <- levels(BodyWeight$Rat)
pintinhos

for (i in pintinhos) {
  dados_i <- BodyWeight[BodyWeight$Rat == i, ]
  dieta_i <- as.character(dados_i$Diet[1])
  
  lines(dados_i$Time, dados_i$weight,
        col = cores[dieta_i])
}

legend("topleft",
       legend = paste("Dieta", names(cores)),
       col = cores,
       lty = 1,
       bty = "n")
```
