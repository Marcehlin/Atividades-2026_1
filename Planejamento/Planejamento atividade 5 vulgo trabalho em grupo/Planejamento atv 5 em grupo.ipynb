## Enunciado:

Busque um conjunto de dados envolvendo pelo mesno 3 fatores. Faça a análise completa destes dados. Escreva seu relatorio em PDF. Prepare uma apresentação de seu Relatorio.

---

Conjunto de dados escolhido: Problem **6-1** do livro "Design and Analysis of Experiments"(5th Edition), Douglas C. Montgomery,

## Contextualização do Problema

Descrição do experimento

Um engenheiro está interessado em investigar os efeitos da:

* velocidade de corte (A),

* geometria da ferramenta (B),

* ângulo de corte (C)

sobre a vida útil de uma ferramenta de máquina, medida em horas.

Foram escolhidos dois níveis para cada fator, e executaram-se três réplicas de um planejamento fatorial completo 2^3.

---

# Objetivo

O objetivo do experimento é identificar quais fatores e interações influenciam significativamente a vida útil da ferramenta.

---

# Estrutura Experimental

- **Fator A:** Taxa de agitação (rpm)
- **Fator B:** Taxa de agitação (rpm)
- **Fator C:** Taxa de agitação (rpm)

- **Variável resposta:** A vida útil de uma ferramenta de máquina, em horas
